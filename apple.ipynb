```python
import pandas as pd
df=pd.read_csv("apple_quality.csv")
df
```

```python
X1=df.drop(columns=['A_id','Acidity'])
X1
```

```python
X1.isna().sum()
```

```python
X1=X1.dropna()
X1
```

```python
X1.isna().sum()
```

```python
gb={'good':1,'bad':0}
# X['Quality']=X['Quality'].map(gb)
X1['Quality']=X1['Quality'].replace(['good','bad'],[1,0])
X1
```

```python
X=X1.drop(columns=['Quality'])
X
```

```python
Y=X1[['Quality']]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,LogisticRegression
```

```python
m1=LinearRegression()
m1.fit(X_train,Y_train)
print(f"Linear Regression Train Score : {m1.score(X_train,Y_train)}")
print(f"Linear Regression Test Score : {m1.score(X_test,Y_test)}")
```

```python
m2=Lasso(alpha=25,max_iter=100,tol=0.1)
m2.fit(X_train,Y_train)
print(f"Lasso Train Score : {m2.score(X_train,Y_train)}")
print(f"Lasso Test Score : {m2.score(X_test,Y_test)}")
```

```python
m3=Ridge(alpha=25,max_iter=100,tol=0.1)
m3.fit(X_train,Y_train)
print(f"Ridge Train Score : {m3.score(X_train,Y_train)}")
print(f"Ridge Regression Test Score : {m3.score(X_test,Y_test)}")
```

```python
m5=LogisticRegression()
m5.fit(X_train,Y_train)
print(f"Logistic Regression Train Score : {m5.score(X_train,Y_train)}")
print(f"Logistic Regression Test Score : {m5.score(X_test,Y_test)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
m4=DecisionTreeClassifier()
m4.fit(X_train,Y_train)
print(f"Decision tree Classifier Train Score : {m4.score(X_train,Y_train)}")
print(f"Decision tree Classifier Test Score : {m4.score(X_test,Y_test)}")
```
